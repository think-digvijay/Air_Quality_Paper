```python
import numpy as np 
import pandas as pd 

import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('E:\\Papers\\Air Quality\\AirQualityUCI.xlsx')
```

```python
print(df.head())
print(df.shape)
```

```python
df.drop('Date', axis=1, inplace=True)
```

```python
#the column seems to have a lot of misplaced data(-200)
df.drop('NMHC(GT)', axis=1, inplace=True)
df.head()
```

```python
#-200 seems to indicate missing data
df.replace(to_replace= -200, value= np.NaN, inplace= True)
```

```python
def VALUE_CORRECTION(col):
    df[col] = df.groupby('Time')[col].transform(lambda x:x.fillna(x.mean()))
```

```python
#filing empty spaces with the mean
col_list = df.columns[1:12]

for i in col_list:
    VALUE_CORRECTION(i)
```

```python
df.info()
```

```python
df.fillna(method='ffill', inplace= True)
```

```python
df.info()
```

```python
# data visualisations
```

```python
#RH vs T
plt.figure(figsize=(25,10))
plt.xlabel('Temperature(C)')
plt.ylabel('Relative Humidity')
plt.title("Relative Humidity vs Temperature(C)")
plt.scatter(df['T'], df['RH'], marker='.', aa=True)
```

```python
#AH vs T
plt.figure(figsize=(25,10))
plt.xlabel('Temperature(C)')
plt.ylabel('Absolute Humidity')
plt.title("Absolute Humidity vs Temperature(C)")
plt.scatter(df['T'], df['AH'], marker='.', aa=True)
```
